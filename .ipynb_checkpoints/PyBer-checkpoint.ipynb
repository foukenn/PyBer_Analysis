```python
# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Files to load
city_data = "Resources/city_data.csv"
ride_data = "Resources/ride_data.csv"
```

```python
# Reads city data and stores it in a pandas dataframe
city_data_df = pd.read_csv(city_data)
city_data_df.head(10)
```

```python
# Reads the ride data and stores
ride_data_df = pd.read_csv(ride_data)
ride_data_df.head(10)
```

```python
# Gets columns & rows that are not null
city_data_df.count()
```

```python
# Get the columns and the rows and ensures there are none null.
city_data_df.isnull().sum()
```

```python
# Gets the data type
city_data_df.dtypes
```

```python
# Gets all unique values
city_data_df["type"].unique()
```

```python
# Get the number of data points from the Urban cities.
sum(city_data_df["type"]=="Rural")
```

```python
# Get the columns and the rows that are not null.
ride_data_df.count()
```

```python
# Get the columns and the rows that are not null.
ride_data_df.isnull().sum()
```

```python
# Get the data types of each column.
ride_data_df.dtypes
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data_df.head()
```

```python
# Creates the urban city dataframe
urbanCities_df = pyber_data_df[pyber_data_df["type"]== "Urban"]
urbanCities_df.head()
```

```python
# Create the Suburban and Rural city DataFrames.
suburbanCities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
ruralCities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
ruralCities_df.head()
```

```python
# Gets the number of rides for urban cities.
urbanRide_count = urbanCities_df.groupby(["city"]).count()["ride_id"]
urbanRide_count.head()
```

```python
# Create the suburban and rural ride count.
suburbanRide_count = suburbanCities_df.groupby(["city"]).count()["ride_id"]

ruralRide_count = ruralCities_df.groupby(["city"]).count()["ride_id"]
```

```python
# Pulls average fare for each city's urban area
urbanAvg_fare = urbanCities_df.groupby(["city"]).mean()["fare"]
urbanAvg_fare.head()
```

```python
# Get average fare for each city in the suburban and rural cities.
suburbanAvg_fare = suburbanCities_df.groupby(["city"]).mean()["fare"]
ruralAvg_fare = ruralCities_df.groupby(["city"]).mean()["fare"]
```

```python
# Get the average number of drivers for each urban city
```

```python
urbanDriver_count = urbanCities_df.groupby(["city"]).mean()["driver_count"]
urbanDriver_count.head()
```

```python
# Get the average number of drivers for each city for the suburban and rural cities.
suburbanDriver_count = suburbanCities_df.groupby(["city"]).mean()["driver_count"]
ruralDriver_count = ruralCities_df.groupby(["city"]).mean()["driver_count"]
```

```python
# Builds urban scatter plot
plt.scatter(urbanRide_count, 
            urbanAvg_fare,
           s = 10*urbanDriver_count, c = "coral",
           edgecolor = "black", linewidths=1,
           alpha = 0.8, label = "Urban")
plt.title("Pyber Ride-Sharing Data (2019)")
plt.ylabel("Average Fare($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
```

```python
# Build the scatter plots for suburban cities.
plt.scatter(suburbanRide_count,
      suburbanAvg_fare,
      s=10*suburbanDriver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
```

```python
# Build the scatter plots for rural cities.
plt.scatter(ruralRide_count,
      ruralAvg_fare,
      s=10*ruralDriver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
```

```python
# Add the scatter charts for each type of city.
plt.scatter(urbanRide_count,
      urbanAvg_fare,
      s=10*urbanDriver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")

plt.scatter(suburbanRide_count,
      suburbanAvg_fare,
      s=10*suburbanDriver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")

plt.scatter(ruralRide_count,
      ruralAvg_fare,
      s=10*ruralDriver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")

# Show the plot
plt.show()
```

```python
#Build the scatter charts for each city type.
plt.subplots(figsize=(10, 6))
plt.scatter(urbanRide_count,
      urbanAvg_fare,
      s=10*urbanDriver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")

plt.scatter(suburbanRide_count,
      suburbanAvg_fare,
      s=10*suburbanDriver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")

plt.scatter(ruralRide_count,
      ruralAvg_fare,
      s=10*ruralDriver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")

# Incorporate the other graph properties
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)



# Create a legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)
#Incororate a tect label about circle size
plt.text(42, 35, "Note: Circle size correlates with driver count per city.", fontsize="12")

# Show the plot
plt.show()
```

```python
#To calculate the percentage of the total fares for each city type based on the total fares, we will:

#Use the groupby() function on the pyber_data_df DataFrame and group by the type of city column so the type of city is the index.
#Apply the sum() method on the fare column to get the Series with the total number of fares for each city type.
#Divide the total fares for each city type by the total of all the fares and multiply by 100.
```

```python
# Gst the sum of the fares for each city type.
sumFares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sumFares_by_type
```

```python
# Gets the sum of all the fares
totalFares = pyber_data_df["fare"].sum()
totalFares
```

```python
# Calculates the percentage of fare for each city type
typePercents = 100 * sumFares_by_type / totalFares
typePercents
```

```python
# Builds the percentage of fares by city type pie chart
#plt.pie(typePercents, labels = ["Rural","Suburban","Urban"])
#plt.show()
```

```python
# Build the percentage of fares by city type pie chart.
plt.pie(typePercents,
    labels=["Rural", "Suburban", "Urban"],
    colors=["gold", "lightskyblue", "lightcoral"],
    explode=[0, 0, 0.1],
    autopct='%1.1f%%',
    shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")
# Show Figure
plt.show()
```

```python
sumDrivers_by_type = pyber_data_df.groupby(["type"]).sum()["driver_count"]
sumDrivers_by_type
totalDrivers = pyber_data_df["driver_count"].sum()
totalDrivers
```

```python
driverPercents = 100 * sumDrivers_by_type / totalDrivers
driverPercents
```

```python
# Build the percentage of fares by city type pie chart.
plt.pie(driverPercents,
    labels=["Rural", "Suburban", "Urban"],
    colors=["gold", "lightskyblue", "lightcoral"],
    explode=[0, 0, 0.1],
    autopct='%1.1f%%',
    shadow=True, startangle=150)
plt.title("% of Total Drivers by City Type")
# Show Figure
plt.show()
```

```python
sumRide_by_type = pyber_data_df.groupby(["type"]).sum()["ride_id"]
sumRide_by_type
totalRides = pyber_data_df["ride_id"].sum()
#totalRides
sumRide_by_type
```

```python
ridePercents = 100 * sumRide_by_type / totalRides
ridePercents
```
